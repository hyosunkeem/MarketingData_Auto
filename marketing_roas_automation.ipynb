# 마케팅 데이터 전처리 자동화

- 주단위로 유입경로별 성과를 트레킹 -> 우리가 할 것

### CPC : Cost Per Click -> 캠페인의 클릭 당 얼마를 지불했는지
### ROAS (Return On Advertising Spend) -> 광고 지출 대비 광고를 통해 벌어들인 수익이 얼마인지

- **정의**: 광고 지출에 대한 수익을 나타내는 마케팅 지표.

- **계산 방법**:
$$
\text{ROAS} = \frac{\text{광고로 인한 수익}}{\text{광고 지출}}
$$

- **중요성**:
  - **높은 ROAS**: 광고 지출 대비 높은 수익을 의미. 광고 캠페인의 효과성을 나타냄.
  - **낮은 ROAS**: 광고 투자에 대한 낮은 수익을 의미. 광고 전략의 개선이 필요함을 시사.


# 주별로 광고 데이터 집계

```python
ad.info()
#date Column이 datetime이 아니라 object로 들어가 있음 -> datetime으로 변경
ad['date'] = pd.to_datetime(ad.date) # 이렇게 하고 꼭 덮어 써주는 걸 해야됨
ad.info() #info를 통해 datetime으로 바뀐게 확인 되었고
ad['week'] = ad.date.dt.to_period('W').dt.to_timestamp() #시작하는 날짜로 설정
ad.tail() #head(), tail() 둘다 함
```

```python
#다음으로 weekly 단위로 campaign의 클릭수와 스팬딩이 얼마나 나왔는지 합산을 해주는 집계
#[['spend', 'clicks']]에서 괄호 두 개 안 했더니 오류 떴었음
# by = [x, y] 두개 위치가 바뀌면 기준도 바뀜
ad_agg = ad.groupby(by = ['campaign', 'week'])[['spend', 'clicks']].sum().reset_index()
```

```python
ad_agg.head()
ad_agg['CPC']= (ad_agg.spend)/(ad_agg.clicks) #CPC : Cost Per Click -> 캠페인의 클릭 당 얼마를 지불했는지
```

```python
ad_agg.tail()
```

```python
#빈값을 organic으로 변경해주고 campaign에 덮어줌
od1['campaign'] = od1.campaign.fillna('organic')
od1.groupby(by = 'campaign').order_amount .sum()
```

```python
#week
od1['week'] = od1.order_date.dt.to_period('W').dt.to_timestamp()
```

```python
od_agg = od1.groupby(by = ['campaign', 'week'])['order_amount'].sum().reset_index()
```

```python
df = pd.merge( od_agg, ad_agg, how = 'left', left_on = ['campaign', 'week'], right_on = ['campaign', 'week'])
df
```

```python
#ROAS 계산 1보다 크면 일단 좋아보이는거, 커질 수록 Business적으로 유리한 듯
(df.order_amount)/(df.spend)
```

```python
#원하는 형태론 전처리중
df['ROAS'] = (df.order_amount)/(df.spend)
df
```

```python
#사실 이 표만 보고 어떤 성과를 분석하기는 어려움이 있는 거 같아서
#피벗 테이블로 좀 가공해봄 > ROAS가 week별로 어떻게 변화했는지 체크 뿅
df.pivot_table(
    index = 'campaign',
    columns = 'week',
    values = 'ROAS'
    )
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 피벗 테이블 생성
pivot_table = df.pivot_table(
    index='campaign',
    columns='week',
    values='ROAS'
)

# 시각화
plt.figure(figsize=(10, 6))

# 캠페인별로 값을 그래프로 그리기
for campaign in pivot_table.index:
    plt.plot(
        pivot_table.columns,   # 주차(week)
        pivot_table.loc[campaign],  # ROAS 값
        marker='o',  # 점 표시
        label=campaign  # 캠페인 레이블
    )

plt.title('ROAS by Campaign and Week', fontsize=16)
plt.xlabel('Week', fontsize=12)
plt.ylabel('ROAS', fontsize=12)
plt.xticks(rotation=45)  # 주차 표시를 기울여 가독성 높임
plt.legend(title='Campaign', fontsize=10)
plt.grid(alpha=0.3)

# 그래프 출력
plt.tight_layout()
plt.show()
```
